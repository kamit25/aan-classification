```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle
```

```python
## read csv

data = pd.read_csv('../Churn_Modelling.csv')
data.head()
```

```python
## processing data,
## drop unnecessary column

data = data.drop(['RowNumber','CustomerId',	'Surname'], axis=1)
data.head()
```

```python
## encode categorical value

label_encoder_gender = LabelEncoder();
data['Gender'] = label_encoder_gender.fit_transform(data['Gender'])
data.head()
```

```python
## onhot encode for Geographic column

from sklearn.preprocessing import OneHotEncoder
onehot_encoder_geo =  OneHotEncoder()
geo_encoder =  onehot_encoder_geo.fit_transform(data[['Geography']])
```

```python
geo_encoder.toarray()
```

```python
onehot_encoder_geo.get_feature_names_out(['Geography'])
```

```python
geo_encoded_df = pd.DataFrame(geo_encoder.toarray(),columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
## concate the one hot encoded column to original data

data= pd.concat([data.drop('Geography',axis=1), geo_encoded_df],axis=1)
data.head()
```

```python
## save the encoders and sscaler
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(label_encoder_gender,file)

with open('onehot_encoder_geo.pkl','wb') as file:
    pickle.dump(onehot_encoder_geo,file)
```

```python
## divide the dataset into dependent and independent features

x = data.drop('Exited', axis=1)
y = data['Exited']

## splits the data in training and testing data

x_train,x_test,y_train,y_test = train_test_split(x,y, train_size=0.2, random_state=42)

## sclae these features

scalar = StandardScaler()
x_train = scalar.fit_transform(x_train)
x_test = scalar.transform(x_test)
```

```python
with open('scalar.pkl', 'wb') as file:
    pickle.dump(scalar,file)
```

## ANN implementation

steps:

1. sequential neural network known as Ann in tensor flow
2. Dense (for hidden layer)
3. Activation function (sigmoid,tahh, relu, leaky relu)
4. optimiser -> backward propogation  -> updating weights
5. loss function
6. metrics -> [Accuracy], [MSE, MAE]
7. training -> logs -> folder -> tensorboard -> visulation. 

```python
import tensorflow as ts
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

```python
print(x_train.shape)
print(x_train.shape[1])
print((x_train.shape[1],))
```

```python
## build ANN model

model = Sequential([
    Dense(64, activation='relu', input_shape = (x_train.shape[1],)), ## hidden layer 1 connected with input layer
    Dense(32, activation='relu') , ## hidden layer 2
    Dense(1, activation='sigmoid') ## output layer
])
```

```python
model.summary()
```

```python
opt = ts.keras.optimizers.Adam(learning_rate=0.01)
loss = ts.keras.losses.BinaryCrossentropy()
loss
```

```python
## compile the model

model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
## setup tensorboard

log_dir = 'logs/fit/' + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tenserflow_callback = TensorBoard(log_dir='log_dir',histogram_freq=1)
```

```python
## setup early stopping

early_stopping_callback = EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
```

```python
### train the model

history=model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100, callbacks=[tenserflow_callback,early_stopping_callback] )
```
